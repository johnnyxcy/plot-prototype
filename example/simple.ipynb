```python
import polars as pl
from mas.libs.phanpy.plotting import Plot, Scatter, Line, factor_cmap
```

```python
data = pl.read_csv("example/students.csv")
data
```

```python
p = (
    Plot(data=data)
    .add(
        Scatter(
            x=pl.col("StudyTimeWeekly"),
            y=pl.col("GPA"),
            line_color=factor_cmap("Gender"),
        )
    )
    .with_facet_wrap("Gender")
)
p
```
