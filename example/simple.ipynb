```python
import polars as pl
from mas.libs.phanpy.plotting import Plot, Scatter, Line, factor_cmap, BoxPlot
```

```python
data = pl.read_csv("example/students.csv")
data
```

```python
BoxPlot(
    data=data,
    y=pl.col("Gender").cast(pl.String),
    x=pl.col("StudyTimeWeekly"),
    fill_color=factor_cmap("ParentalEducation"),
).with_y_axis(typ="categorical", factors=["0", "1"]).with_legend_layout(
    placement="right"
)
```
